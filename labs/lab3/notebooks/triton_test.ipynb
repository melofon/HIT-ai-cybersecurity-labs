```python
import os
import tritonclient.http as httpclient

#triton_url = os.environ.get("TRITON_HTTP_URL", "triton:8000")
triton_url = "triton:8000"
print("TRITON_HTTP_URL:", triton_url)

client = httpclient.InferenceServerClient(url=triton_url, verbose=True)

print("Server live :", client.is_server_live())
print("Server ready:", client.is_server_ready())

server_metadata = client.get_server_metadata()
print(f"server metadata: {server_metadata}")

repo_index = client.get_model_repository_index()
print(f"list of models: {repo_index}")
```
